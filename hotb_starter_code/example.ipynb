```python
import mne
import os

# --- 1. Define the path to your .fif file ---
sample_data_folder = "data"
sample_data_raw_file = os.path.join(sample_data_folder, '20251128_0010-raw.fif')

# --- 2. Load the data ---
# mne.io.read_raw_fif() reads the .fif file.
# preload=True loads the data into memory for easier processing.
raw = mne.io.read_raw_fif(sample_data_raw_file, preload=True)

# --- 3. Explore the data ---
# Print basic information about the Raw object
print(raw)
print(raw.info)

# --- 4. Visualize the data ---
# Plot the power spectral density (PSD) of the data
# This helps to see the frequency content and identify potential noise.
raw.plot_psd(fmax=125)

# Plot the raw sensor data interactively
# You can scroll through time and channels.
# raw.apply_function(lambda x: x*1e-6)
raw.plot(duration=5, n_channels=4, scalings='auto')
```
